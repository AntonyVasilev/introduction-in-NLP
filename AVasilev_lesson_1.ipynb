```python
train_df  = pd.read_csv('train_tweets.csv', encoding='utf-8')
test_df = pd.read_csv('test_tweets.csv', encoding='utf-8')
```

```python
df = train_df.append(test_df, ignore_index = True, sort = False)
df.head()
```

```python
# убираю @user
def drop_user(text):
    return re.sub('@[\w]*', '', text)
```

```python
df.tweet = df.tweet.apply(drop_user)
```

```python
# привожу к нижнему регистру
def make_lower(text):
    return text.lower()
```

```python
df.tweet = df.tweet.apply(make_lower)
```

```python
def replace_text(text, dictionary):
    result = [dictionary[word] if word in dictionary.keys() else word
                 for word in text.split()]
    return ' '.join(result)
```

```python
# заменяю сокращения с апострофами полными формами
df.tweet = df.tweet.apply(lambda x: replace_text(x, apostrophe_dict))
```

```python
# заменяю сокращения полными формами
df.tweet = df.tweet.apply(lambda x: replace_text(x, short_word_dict))
```

```python
# заменяю эмотиконы словами
df.tweet = df.tweet.apply(lambda x: replace_text(x, emoticon_dict))
```

```python
# удаляю пунктуацию
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)
```

```python
df.tweet = df.tweet.apply(remove_punctuation)
```

```python
# удаляю спецсимволы
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9]', ' ', text)
```

```python
df.tweet = df.tweet.apply(remove_special_characters)
```

```python
# удаляю числа
def remove_numbers(text):
    return re.sub(r'[^a-zA-Z]', ' ', text)
```

```python
df.tweet = df.tweet.apply(remove_numbers)
```

```python
# удаляю из текста слова длиной в 1 символ
def remove_short_words(text):
    return ' '.join([word for word in text.split() if len(word) > 1])
```

```python
df.tweet = df.tweet.apply(remove_short_words)
```

```python
# делю твиты на токены
df['tweet_token'] = df.tweet.apply(lambda x: nltk.tokenize.word_tokenize(x))
```
