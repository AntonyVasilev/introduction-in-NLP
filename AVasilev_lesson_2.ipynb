```python
import pandas as pd
import numpy as np
# import re
# import matplotlib.pyplot as plt
# import seaborn as sns
# import nltk
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
# import os
import pickle

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
with open('combine_df.pickle', 'rb') as f:
    combine_df = pickle.load(f)
```

```python
combine_df = combine_df.set_index('id')
combine_df.head()
```

### 1. Использую CountVectorizer для создания мешков слов

```python
c_vec = CountVectorizer(max_df=0.9, max_features=1000, stop_words='english')
```

#### 1) 'tweet_stemmed'

```python
bow_tweet_stemmed_c_vec = c_vec.fit_transform([' '.join(sent) for sent in combine_df.tweet_stemmed])
# Обображаю Bag-of-Words как DataFrame
pd.DataFrame(bow_tweet_stemmed_c_vec.toarray(), columns = c_vec.get_feature_names()).sample(5)
```

#### 2) 'tweet_lemmatized'

```python
bow_tweet_lemmatized_c_vec = c_vec.fit_transform([' '.join(sent) for sent in combine_df.tweet_lemmatized])
# Обображаю Bag-of-Words как DataFrame
pd.DataFrame(bow_tweet_lemmatized_c_vec.toarray(), columns = c_vec.get_feature_names()).sample(5)
```

### 2. Использую CountVectorizer для создания мешков слов

```python
tf_vec = TfidfVectorizer(max_df=0.9, max_features=1000, stop_words='english')
```

#### 1) 'tweet_stemmed'

```python
bow_tweet_stemmed_tf_vec =tf_vec.fit_transform([' '.join(sent) for sent in combine_df.tweet_stemmed])
# Обображаю Bag-of-Words как DataFrame
pd.DataFrame(bow_tweet_stemmed_tf_vec.toarray(), columns = tf_vec.get_feature_names()).sample(5)
```

#### 2) 'tweet_lemmatized'

```python
bow_tweet_lemmatized_tf_vec = tf_vec.fit_transform([' '.join(sent) for sent in combine_df.tweet_lemmatized])
# Обображаю Bag-of-Words как DataFrame
pd.DataFrame(bow_tweet_lemmatized_tf_vec.toarray(), columns = tf_vec.get_feature_names()).sample(5)
```

### 3.

```python
# Загружаю данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаю датафрейм
corpus_df = pd.DataFrame()
corpus_df['text'] = texts
corpus_df['label'] = labels
corpus_df.head(5)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(corpus_df['text'], corpus_df['label'])

# labelEncode целевую переменную
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_valid = encoder.fit_transform(y_valid)
```

```python
# CountVectorizer
X_train_count_vec =  c_vec.transform(X_train)
X_valid_count_vec =  c_vec.transform(X_valid)

clf_count = LogisticRegression(max_iter=200)
clf_count.fit(X_train_count_vec, y_train)
preds_count = clf_count.predict(X_valid_count_vec)

accuracy_score(y_valid, preds_count)
```

```python
# CountVectorizer
X_train_tfidf_vec = tf_vec.transform(X_train)
X_valid_tfidf_vec = tf_vec.transform(X_valid)

clf_tfidf = LogisticRegression()
clf_tfidf.fit(X_train_tfidf_vec, y_train)
preds_tfidf = clf_tfidf.predict(X_valid_count_vec)

accuracy_score(y_valid, preds_tfidf)
```
